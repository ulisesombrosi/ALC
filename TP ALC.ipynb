```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Ejercicio 1
def metodoPotencia(A, v, k):
    a = []
    for i in range(k):
        v = (A @ v)/np.linalg.norm(A @ v)
        lambdas = (v @ A @ v)/(v @ v)
        a.append(lambdas)
    return(a)

#A = np.array([[0.9,0.075,.025],[.15,.8,.05],[.25,.25,.5]])
#v = np.array([1,2,3])
A = np.array([[-2,0,0],[0,-5,6],[0,-3,4]])
v = np.array ([0,1,1])

print(metodoPotencia(A,v,1))
print(metodoPotencia(A,v,10))
print(metodoPotencia(A,v,1000))
```

```python
#Ejercicio 2
def matriz(t):
    A = np.random.rand(t, t)
    
    B = np.random.rand(t,t)
    for i in range (0,t):
        for j in range (i,t):
            B[j][i] = B[i][j]
    
    
    C = np.copy(B)
    for i in range (0,t):
        C[i][i] = B[i][i] + 100
        
    D = np.copy(B)
    for i in range (0,t):
        D[i][i] = B[i][i] + 1000
    return(A,B,C,D)


A,B,C,D = matriz(100)
print(A)
print(B)
print(C)
print(D)
```

```python
#Inciso 2a
rango = list(range(1,101))

plt.plot(rango,metodoPotencia(A, rango, 100), color= "green")
plt.plot(rango,metodoPotencia(B, rango, 100), color = "red")
plt.plot(rango,metodoPotencia(C, rango, 100), color= "blue")
plt.plot(rango,metodoPotencia(D, rango, 100), color = "black")
```

Inciso 2b:

Al comparar los gráficos, podemos observar saltos abruptos entre la matriz D, la matriz C y las demás. Esto se debe a que, mientras que la matriz A y B contienen elementos entre 0 y 1; las matrices C y D fueron definidas a partir de la matriz B, sumándoles 100 o 1000 respectivamente a los valores de la diagonal.
